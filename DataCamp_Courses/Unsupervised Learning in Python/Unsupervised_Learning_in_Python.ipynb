### Inertia measures clustering quality
* Measures how spread out the clusters are (<i>lower</i> is better).
* Distance from each sample to centroid of its cluster
* After fit() method, inertia is available as an attribute inertia_

```python
from sklearn.cluster import KMeans

df = pd.read_csv('data/iris_prepared.csv')

df = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

model = KMeans(n_clusters=3)
model.fit(df)
print(model.inertia_)
```

### How many clusters?

```python
ks = range(1, 6)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters = k)
    
    # Fit model to samples
    model.fit(df)
    
    # Append the inertia to the list of inertias
    inertia = model.inertia_
    inertias.append(inertia)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
